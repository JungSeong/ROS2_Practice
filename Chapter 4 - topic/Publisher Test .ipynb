# 토픽 발행을 위한 test_node의 작성

/turtle1/cmd_vel라는 토픽의 데이터 타입이 geometry_msgs/msg/Twist 이다.

```python
import rclpy as rp
from geometry_msgs.msg import Twist

rp.init()
test_node = rp.create_node('pub_test') #pub_test라는 노드의 이름을 코드 상에서는 test_node로 지정
```

```python
msg = Twist()
print(msg)
```

```python
msg.linear.x = 2.0
print(msg)
```

```python
msg.linear.y=4.0
print(msg)
```

```python
pub = test_node.create_publisher(Twist, '/turtle1/cmd_vel', 10)
pub.publish(msg)
```

```python
msg.angular.z = 2.0
msg.linear.y=0.0
print(msg)
```

```python
pub = test_node.create_publisher(Twist, '/turtle1/cmd_vel', 10)
pub.publish(msg)
```

```python
msg.linear.x = 4.0
```

```python
cnt = 0

def timer_callback():
    global cnt
    cnt += 1
    
    print(cnt)
    pub.publish(msg)

    if cnt>3 :
        raise Exception("publisher stop")
```

test_node.create_timer(timer_period, timer_callback) : timer_period라는 주기 마다 timer_callback 이라는 함수를 실행

timer_period가 0.1 : 0.1초 마다 timer_callback을 실행

```python
timer_period = 0.1
timer = test_node.create_timer(timer_period, timer_callback)
rp.spin(test_node)
```

```python
test_node.destroy_node()
```
